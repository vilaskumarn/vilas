**Strings Hands-On Session**

**Concatination**

```python
fname="Vilas"
lname="Kumar"

fullname=fname+" "+lname

print(f"Your name is : {fullname}")
```

**Repetition**

```python
r=10
fname=" Vilas"
print(f"Your name is now repeated :{r}\n {fname*r} ")
```

```python
print(f"{"="*r}\n{fullname}\n{"="*r}")
```

```python
print(f"Length of First name {fname} is: {len(fname)}\nLength of Last name {lname} is: {len(lname)}")
```

**Indexing & Slicing**

```python
print(f"{fullname[0]}")
```

```python
print(f"{fullname[:5]}")
```

```python
print(f"{fullname[6:]}")
```

```python
print(f"{fullname[0:6:2]}")
```

```python
print(f"{fullname[0:6:4]}")
```

**Lower & Upper Function**

```python
print(f"First name in complete UpperCase : {fname.upper()}")
```

```python
print(f"First name in complete LowerCase : {fname.lower()}")
```

**Strip Function**

```python
data=" Data Science  "
print(f"Data at present :{data}\n Data after Striping :{data.strip()}")
```

```python
data=" Data Science  "
print(f"Data at present :{data}\n Data after Left Striping :{data.lstrip()}")
```

```python
data=" Data Science  "
print(f"Data at present :{data}\n Data after Right Striping :{data.rstrip()}")
```

```python
data=" Data Science=="
print(f"Data at present :{data}\n Data after Striping :{data.strip("=")}")
```
