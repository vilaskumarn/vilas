# Views vs copies
This is one of the most important beginner-to-intermediate topics. A slice usually creates a view, not an independent copy.

```python
base = np.array([1, 2, 3, 4, 5])
view = base[1:4]
view
```

```python
view[0]=999
print(view)
```

```python
copy=base[1:4].copy()
copy
```

```python
copy[0]=-100
copy
```

# Reshaping arrays
Reshaping appears everywhere in data preprocessing and ML input preparation.

```python
arr = np.arange(12)
arr
```

```python
reshaped = arr.reshape(3, 4)
reshaped
```

```python
reshaped=arr.reshape(2,6)
reshaped
```

```python
reshaped=arr.reshape(4,3)
reshaped
```

```python
new_reshape=reshaped.reshape(1,12)
new_reshape
```

```python
flattened = reshaped.flatten()
flattened
```

**order (optional) — specifies the reading order of elements:**

- 'C' → Row-major (C-style) order (default) — read row by row.
- 'F' → Column-major (Fortran-style) order — read column by column.
- 'A' → Fortran order if array is Fortran-contiguous, else C order.
- 'K' → Flatten in the order elements occur in memory.

```python
reshaped
```

```python
flattened = reshaped.flatten('C')
flattened
```

```python
reshaped
```

```python
flattened = reshaped.flatten('F')
flattened
```

```python
flattened = reshaped.flatten('A')
flattened
```

```python
flattened = reshaped.flatten('K')
flattened
```

The .ravel() function in NumPy is used to return a contiguous flattened array from a given multi-dimensional array. This means it converts the input array into a one-dimensional array while maintaining the same data type. A copy of the data is made only if necessary, making it efficient in terms of memory usage.

```python
raveled = reshaped.ravel()
raveled
```

```python
reshaped
```

In NumPy, the .T attribute is used to transpose an array, which means flipping its axes. For example, rows become columns and vice versa. However, .T does not create a copy of the array; instead, it returns a view of the original array with transposed axes.

```python
transposed = reshaped.T
transposed
```

# Combining and splitting arrays
These operations are useful when engineering features or preparing train/test style structures manually.

numpy.hstack stacks arrays horizontally, meaning it appends arrays along the columns. This is particularly useful when you want to combine arrays side-by-side. For hstack to work, the arrays must have the same number of rows.

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
```

```python
v=np.vstack([a,b])
v
```

```python
h=np.hstack([a,b])
h
```

```python
split_v=np.vsplit(v,2)
split_v
```

```python
split_v=np.vsplit(v,4)
split_v
```

```python
split_h=np.hsplit(h,2)
split_h
```

```python
split_h=np.hsplit(v,2)
split_h
```
