```python
# Question 49: Convert "hello world" to ROT13 encoding
print("\nQuestion 49: Convert 'hello world' to ROT13 encoding")
```

```python
string=input("Enter a word to convert to ROT13 encoding :")
print(f"The ROT13 encoding of {string} is : ",end='')
for i in string.upper():
    if ord(i)<78 and ord(i)>64:
        print(chr(ord(i)+13),end='')
    elif ord(i)>77 and ord(i)<91:
        print(chr(ord(i)-13),end='')
    else:
        print(i,end='')
```
