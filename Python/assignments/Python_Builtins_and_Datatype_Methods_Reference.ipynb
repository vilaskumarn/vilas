## 1. Built-in Functions — Global Essentials

These functions are not methods of one datatype only. They work across many Python objects.

```python
# type(), isinstance(), id(), callable(), dir()
x = [1, 2, 3]
print(type(x))
print(isinstance(x, list))
print(id(x))
print(callable(len))
print('append' in dir(x))
```

```python
# abs(), round(), pow(), divmod(), sum(), min(), max()
print(abs(-25))
print(round(3.14159, 3))
print(pow(2, 5))
print(divmod(17, 5))   # (quotient, remainder)
print(sum([1, 2, 3, 4]))
print(min([9, 4, 7]))
print(max([9, 4, 7]))
```

```python
# len(), sorted(), reversed(), enumerate(), zip(), any(), all()
nums = [5, 2, 9, 1]
print(len(nums))
print(sorted(nums))
print(list(reversed(nums)))
print(list(enumerate(['a', 'b', 'c'], start=1)))
print(list(zip(['a', 'b'], [10, 20])))
print(any([0, '', None, 5]))
print(all([1, True, 'hello']))
```

```python
# map(), filter(), iter(), next()
nums = [1, 2, 3, 4, 5]
print(list(map(lambda x: x**2, nums)))
print(list(filter(lambda x: x % 2 == 0, nums)))

it = iter(nums)
print(next(it))
print(next(it))
```

```python
# Conversion constructors
print(int('123'))
print(float('3.14'))
print(str(999))
print(list('python'))
print(tuple([1, 2, 3]))
print(set([1, 2, 2, 3]))
print(dict([('a', 1), ('b', 2)]))
print(bool(1))
print(complex(2, 3))
print(range(5))
```

```python
# chr(), ord(), bin(), oct(), hex(), format(), repr(), hash()
print(chr(65))
print(ord('A'))
print(bin(10))
print(oct(10))
print(hex(255))
print(format(12.3456, '.2f'))
print(repr('hello
world'))
print(hash('python'))
```

Important built-ins and operations used with integers.

```python
x = 25
print(type(x))
print(abs(-x))
print(pow(x, 2))
print(divmod(25, 4))
print(bin(x), oct(x), hex(x))
print(x.bit_length())   # number of bits needed
print((10).to_bytes(2, byteorder='big'))
print(int.from_bytes(b' 
', byteorder='big'))
```

## 3. Float (`float`)

```python
x = 3.14159
print(type(x))
print(round(x, 2))
print(float('2.718'))
print(x.is_integer())
print((15.0).is_integer())
print(x.as_integer_ratio())
print(x.hex())
print(float.fromhex(x.hex()))
```

## 6. String (`str`) — Most Important Methods

```python
s = '  hello python world  '
print(len(s))
print(s.upper())
print(s.lower())
print(s.title())
print(s.capitalize())
print(s.swapcase())
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s = 'hello python world'
print(s.split())
print('-'.join(['data', 'science']))
print(s.replace('python', 'AI'))
print(s.find('python'))
print(s.index('python'))
print(s.count('o'))
print(s.startswith('hello'))
print(s.endswith('world'))
```

```python
# String validation methods
print('python'.isalpha())
print('12345'.isdigit())
print('12345'.isnumeric())
print('abc123'.isalnum())
print('hello'.islower())
print('HELLO'.isupper())
print('Hello World'.istitle())
print('   '.isspace())
```

```python
# Encoding / formatting / partitioning
s = 'data science'
print(s.center(20, '-'))
print(s.ljust(20, '.'))
print(s.rjust(20, '.'))
print(s.partition(' '))
print('name={}, score={}'.format('Vilas', 95))
print('python'.encode('utf-8'))
```

### Must-know `str` methods
- Case conversion: `upper()`, `lower()`, `title()`, `capitalize()`, `swapcase()`
- Cleanup: `strip()`, `lstrip()`, `rstrip()`
- Search: `find()`, `index()`, `count()`, `startswith()`, `endswith()`
- Split / join: `split()`, `join()`, `partition()`
- Replace / format: `replace()`, `format()`
- Validation: `isalpha()`, `isdigit()`, `isnumeric()`, `isalnum()`, `islower()`, `isupper()`, `isspace()`

## 7. List (`list`) — Most Used in Python

```python
nums = [10, 20, 30]
nums.append(40)
nums.extend([50, 60])
nums.insert(1, 15)
print(nums)

nums.remove(15)
popped = nums.pop()
print(nums)
print('Popped:', popped)
```

```python
nums = [40, 10, 30, 20, 10]
print(nums.index(30))
print(nums.count(10))
nums.sort()
print(nums)
nums.reverse()
print(nums)
copy_nums = nums.copy()
print(copy_nums)
nums.clear()
print(nums)
```

### Must-know `list` methods
- `append()`
- `extend()`
- `insert()`
- `remove()`
- `pop()`
- `index()`
- `count()`
- `sort()`
- `reverse()`
- `copy()`
- `clear()`

## 9. Set (`set`)

```python
a = {1, 2, 3}
b = {3, 4, 5}

a.add(10)
a.update([20, 30])
print(a)
a.remove(10)
a.discard(999)  # no error
print(a)

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(a.symmetric_difference(b))
```

```python
a = {1, 2, 3}
b = {1, 2, 3, 4, 5}
print(a.issubset(b))
print(b.issuperset(a))
print(a.isdisjoint({7, 8}))

x = {1, 2}
y = {2, 3}
x.intersection_update(y)
print(x)

y.clear()
print(y)
```

### Must-know `set` methods
- `add()`
- `update()`
- `remove()`
- `discard()`
- `pop()`
- `clear()`
- `union()`
- `intersection()`
- `difference()`
- `symmetric_difference()`
- `issubset()`
- `issuperset()`
- `isdisjoint()`
- `intersection_update()` and other in-place update methods

## 10. Frozenset (`frozenset`)

```python
fs = frozenset([1, 2, 3, 3])
print(fs)
print(fs.union({4, 5}))
print(fs.intersection({2, 3, 9}))
print(fs.issubset({1, 2, 3, 4}))
```

### `frozenset` notes
- Immutable version of set
- Supports read-only set operations like `union()`, `intersection()`, `issubset()`
- Can be used as a dictionary key

## 11. Dictionary (`dict`) — Essential for Data Science

```python
student = {'name': 'Vilas', 'age': 22, 'city': 'Bengaluru'}
print(student.keys())
print(student.values())
print(student.items())
print(student.get('name'))
print(student.get('salary', 0))

student.update({'age': 23, 'score': 95})
print(student)
print(student.pop('city'))
print(student)
```

```python
d = {'a': 1, 'b': 2}
print(d.setdefault('c', 3))
print(d)
print(d.popitem())
print(d)
copy_d = d.copy()
print(copy_d)

d.clear()
print(d)

# class methods
print(dict.fromkeys(['x', 'y', 'z'], 0))
```

### Must-know `dict` methods
- `keys()`
- `values()`
- `items()`
- `get()`
- `update()`
- `pop()`
- `popitem()`
- `setdefault()`
- `copy()`
- `clear()`
- `fromkeys()`

## 12. Range (`range`)

```python
r = range(1, 10, 2)
print(r)
print(list(r))
print(len(r))
print(r.start, r.stop, r.step)
print(5 in r)
print(r.index(5))
print(r.count(5))
```

### `range` supports
- iteration
- membership testing
- `count()`
- `index()`
- attributes: `.start`, `.stop`, `.step`

## 13. Bytes (`bytes`) and Bytearray (`bytearray`)

```python
b = bytes('hello', 'utf-8')
ba = bytearray(b)
print(b)
print(ba)
print(b.decode('utf-8'))
print(b.startswith(b'he'))
print(b.upper())

ba[0] = 72   # mutable
print(ba)
ba.append(33)
print(ba)
```

### `bytes` methods are similar to string methods on binary data
### `bytearray` adds mutability with methods like `append()`, item assignment, etc.

## 14. Useful Built-ins for Iterables & Functional Style

```python
data = [1, 2, 3, 4, 5]

print(list(map(lambda x: x * 10, data)))
print(list(filter(lambda x: x % 2 == 1, data)))
print(list(enumerate(data)))
print(list(zip(data, 'abcde')))
print(sum(data))
print(any(x > 4 for x in data))
print(all(x > 0 for x in data))
```

## 15. Built-ins Commonly Used with Strings, Lists, Dicts, Sets in Interviews

```python
words = ['data', 'science', 'python', 'ai']
print(sorted(words, key=len))
print(max(words, key=len))
print(min(words, key=len))
print(list(reversed(words)))

scores = {'Alice': 88, 'Bob': 95, 'Carol': 72}
print(sorted(scores.items(), key=lambda x: x[1], reverse=True))
```
