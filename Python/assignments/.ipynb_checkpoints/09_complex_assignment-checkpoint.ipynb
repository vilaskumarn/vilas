```python
# Question 10: Find the phase angle of complex number 1 + 1j
print("\nQuestion 10: Find the phase angle of complex number 1 + 1j")
# Your code here 
```

```python
import cmath

a=1 + 1j
ph=cmath.phase(a)
res=ph*180/3.14
round(res,3)
```

```python
import numpy as np
a=1 + 1j
phase=np.angle(a)
res=phase*180/3.14
round(res,3)
```
