**2. Create a dictionary and iterate over its keys and values.**

```python
student = {
    "name": "Vilas",
    "age": 23,
    "address": {
        "city": "Bengaluru",
        "pincode": 560049
    }
}
```

```python
for key,value in student.items():
    print(key,value)
```

```python
for key,value in student.items():
    if type(value) is dict:
        print(key+":")
        for i in value.keys():
            print(i,value[i])
    else:
        print(key,value)
```
