# Questions On List, Tuples,sets

**1. Create a list and find the sum of all its elements.**

```python
l=[1,5,3,7,8,9]
res=0
for i in range(len(l)):
    res+=l[i]
print(f"Sum of list items : {res}")
```

**2. Write a program to find the maximum and minimum in a list.**

```python
l=[1,5,3,7,8,9]
maxn=0
minn=l[0]
for i in range(len(l)):
    if maxn<l[i]:
        maxn=l[i]
    if minn>l[i]:
        minn=l[i]
print(f"Max Number is : {maxn} \nMin Number is : {minn}")
```

**3. Create a program that removes duplicates from a list.**

```python
l = [0,1,2,2,2,5,3,7,5,8,7,8,2,0,9,1]
l.sort()

unique = []
for num in l:
    if num not in unique:
        unique.append(num)

print(f"Number: {unique}")
```

**4. Create a tuple and access its elements.**

```python
t=(1,5,3,7,8,9)
for i in range(len(t)-1):
    print(f"At position {i} : {t[i]}")
```

**5. Convert a list into a tuple and vice versa.**

```python
l=[1,5,3,7,8,9]
t=(5,4,1,4,5,6,9)
ct=[]
cl=tuple(l)
for i in t:
    ct.append(i)
print(f"list before converting to tuple : {l}\nAfter conversion : {cl}")
print(f"Tuple before converting to list : {t}\nAfter conversion using constructor : {ct}")
```

**6. Write a program to sort a list in ascending order.**

```python
l = [5, 2, 9, 1, 7]

for i in range(len(l)):
    for j in range(0, len(l) - i - 1):
        if l[j] > l[j + 1]:
            l[j], l[j + 1] = l[j + 1], l[j]

print("Sorted list:", l)
```
