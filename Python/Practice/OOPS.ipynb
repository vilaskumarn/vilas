```python
class BankAccount:
    # Constructor - initializes the object
    def __init__(self, account_number, account_holder, balance):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    # Deposit method
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: ${amount}. New balance: ${self.balance}")

    # Withdraw method
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrawn: ${amount}. New balance: ${self.balance}")
        else:
            print("Insufficient funds!")

    # Check balance method
    def check_balance(self):
        print(f"Current balance: ${self.balance}")
```

## Exercise 1: Movie Class

Create a `Movie` class with:

- **Attributes:**
  - `title` → Name of the movie (string)
  - `director` → Director of the movie (string)
  - `year` → Release year (integer)
  - `rating` → Movie rating (float, e.g., out of 10)

- **Methods:**

  - `display_info()`  
    - Prints or returns all movie details in a readable format  
    - Example output:  
      `"Inception (2010), directed by Christopher Nolan, Rating: 8.8"`

  - `update_rating(new_rating)`  
    - Takes a new rating as input  
    - Updates the movie's `rating` attribute  
    - Optionally validate (e.g., rating should be between 0 and 10)

  - `is_classic()`  
    - Checks if the movie was released before the year 2000  
    - Returns `True` if `year < 2000`, otherwise `False`

```python
class Movie:

    def __init__(self,title,director,year,rating):
        self.title=title
        self.director=director
        self.year=year
        self.rating=rating

    def display_info(self):
        print(f"{self.title} ({self.year}), directed by {self.director},Rating : {self.rating} out of 10")
        
    def update_rating(self,new_rating):
        self.rating=new_rating
        if self.rating<0 or self.rating>10:
            print("Rating should be between 0 to 10")
        print(f"New rating for {self.title} is {self.rating}")
        
    def is_classic(self):
        if self.year<2000:
            return True
        else:
            return False
```
