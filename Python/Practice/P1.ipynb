**Calculator**

```python
a=int(input("Enter First number:"))
b=int(input("Enter Second number:"))

print(f"Operations :\n 1.Addition --> '+' \n 2.Subtraction --> '-' \n 3.Multiplication -->'*' \n 4.Division -->'/'")
op=input("Enter the operation symbol:")
match op:
    case '+':
        res=a+b
    case '-':
        res=a-b
    case '*':
        res=a*b
    case '/':
        if b==0:
            print(f"Denominator can't be zero ('0') to perform division")
        else:
            res=a/b
    case _:
        print(f"{op} is an invalid operation")

print(f"{a} {op} {b} = {res}")
```

**Guess the number Game**

```python
import random as rd
number=rd.randint(1,100)
atempts=7
print("Hello!, Welcome To Guess the Number Game. \n Computer Generates a secret number between '1 to 100' \n You'll be given '7' attempts to try guessing the number. \n All the Best!!")
for i in range(1,atempts+1):
    a=int(input(f"Attempt {i}/{atempts} : Enter Your Guess:"))
    if a==number:
        print(f"Congrats your guess {a} is correct!!!")
        break
    elif a<1 or a>100:
        print("Enter a number between '1 to 100' ")
    elif a>number:
        print("Too high!!, Try a small number")
    else:
        print("Too small!!, Try a large number")
else:
    print(f"Out of attempts!! The number was : {number}")
    
    
```

*Printing Prime Numbers between '1 to 100'* 

```python
print("Prime Numbers between 1 to 100")
for i in range(2,101):
    is_prime=True

    for j in range(2,int(i**0.5)+1):
        if i%j==0:
            is_prime=False
            break
    if is_prime:
        print(f"{i} ")
```

**Fibonacci Series**

```python
a=int(input("Enter the nth term :"))
series=[]
if a<1:
      print("Enter a valid number")
elif a==1:
    series.append(0)
elif a==2:
    series=[0,1]
else:
    series=[0,1]
    for i in range(2,a):
        new=series[i-1]+series[i-2]
        series.append(new)
print(f"Fibonacci Series are : {series}")
```

**GCD Program**

```python
a=int(input("Enter First number:"))
b=int(input("Enter Second number:"))

def gcdpr(a,b):
    if a==0:
        return b
    if b==0:
        return a
    if a%2==0 and b%2==0:
        return 2*gcdpr(a//2,b//2)
    if a%2==0:
        return gcdpr(a//2,b)
    if b%2==0:
        return gcdpr(a,b//2)
    return gcdpr(abs(a-b),min(a,b))
print(gcdpr(a,b))
    
```
