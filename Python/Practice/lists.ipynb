**Task 3: Insert at Specific Position**\
numbers = [1, 2, 4, 5]

Insert 3 in the correct position.

```python
numbers = [1, 2, 4, 5]
numbers.insert(2,3)
print(numbers)
```

**Task 4: Remove an Item**\
fruits = ["apple", "banana", "mango", "grapes"]

Remove "mango".

```python
fruits = ["apple", "banana", "mango", "grapes"]
fruits.remove("mango")
print(fruits)
```

**Task 5: Remove Last Item**\
items = ["pen", "book", "eraser"]

Remove the last item.

```python
items = ["pen", "book", "eraser"]
items.pop() # or items.pop(-1)
print(items)
```

**Task 6: Find Index**\
animals = ["cat", "dog", "lion", "tiger"]

Find the position of "lion".

```python
animals = ["cat", "dog", "lion", "tiger"]
print(animals.index("lion"))
```

**Task 7: Count Occurrences**\
nums = [1, 2, 2, 3, 2, 4]

Count how many times 2 appears.

```python
nums = [1, 2, 2, 3, 2, 4]
print(nums.count(2))
```

**Task 8: Sort the List**\
marks = [85, 45, 90, 60]

Sort the list in ascending order.

```python
marks = [85, 45, 90, 60]
marks.sort()
print(marks)
```

**Task 9: Reverse the List**\
letters = ["a", "b", "c", "d"]

Reverse the list.

```python
letters = ["a", "b", "c", "d"]
letters.reverse()
print(letters)
```

**Task 10: Copy the List** \

**Task 11: Clear the List**\
data = [1, 2, 3, 4]

Remove all elements from the list.

```python
data = [1, 2, 3, 4]
data.clear()
print(data)
```

**Task 12: Extend a List**\
list1 = [1, 2]\
list2 = [3, 4]

Add all elements of list2 into list1.

```python
list1 = [1, 2]
list2 = [3, 4]
list1.extend(list2)

print(list1)
```
