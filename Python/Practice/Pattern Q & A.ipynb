**Count Even numbers in nested loop**

```python
data = [
    [2,4,6],
    [2,3,5],
    [7,8,9]
]
count=0

for i in data:
    for j in i:
        if j%2==0:
            count+=1
print(f"There are {count} Even numbers in the nested list : {data}")
```
