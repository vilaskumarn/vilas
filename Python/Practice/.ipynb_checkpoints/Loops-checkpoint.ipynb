## For Loop

**1. Reverse for loop. Print N to 1**

```python
a=int(input("Enter the end number :"))
for i in range(a,0,-1):
    print(i)
```

**2. Sum of N terms**

```python
a=int(input("Enter the end number :"))
res=0
for i in range(1,a+1):
    # res+=i 
    res=res+i
    # print(res)
print(f"Sum of {a} terms is : {res}")
```

**3. Sum of Even & Odd numbers of N terms, Separately**

```python
a=int(input("Enter the end number :"))
even=0
odd=0
for i in range(1,a+1):
    if i%2==0:
        even+=i
    else:
        odd+=i

print(f"Sum of Even Numbers : {even} \n Sum of Odd Numbers : {odd}")
```

**4. Print all factors of a number**

```python
a=int(input("Enter the end number :"))
print(f"All Factors of {a} are :")
for i in range(1,a+1):
    if a%i==0:
        print(i)
```

**5. Perfect Number. Sum of factors = original number**

```python
a=int(input("Enter the end number :"))
res=0
for i in range(1,a):
    if a%i==0:
        res+=i
if res==a:
    print(f"{a} is a Perfect Number!")
else:
    print(f"{a} is not a Perfect Number!")
```

**7. Count of alphabets,special symbols and numbers in a given string**

```python
a=input("Enter String:")
al=0
sc=0
num=0
for i in range(0,len(a)):
    b=ord(a[i])
    if 90>=b>=65 or 122>=b>=97:
        al+=1
    elif 57>=b>=48:
        num+=1
    else:
        sc+=1
        
print(f"Count of Alphabets : {al} \nCount of Special Characters : {sc} \nCount of Numbers : {num}")
```

**4. Amstrong Numbers**

```python
n=int(input("Enter a number :"))
t=n
l=[]
while n>0:
    l.append(n%10)
    n=n//10
s=len(l)
res=0
for i in l:
    res+=i**s
if res==t:
    print(f"{t} is an Amstrong Number")
else:
    print(f"{t} is not an Amstrong Number")
```

**5. Printing Amstrong numbers upto N terms**

```python
n=int(input("Enter a number :"))

for i in range(1,n+1):
    j=i
    l=[]
    while j>0:
        l.append(j%10)
        j=j//10
    s=len(l)
    res=0
    for k in l:
        res+=k**s
    if res==i:
        print(f"{i} is an Amstrong Number")
```

**6. Write a Python program to print a pattern of stars in a triangle.**

```python
n=int(input("Enter the size of triangle"))

for i in range(1,n+1):
    print(f"{i*'*'}")
```

```python
n=int(input("Enter the size of triangle"))

for i in range(1,n+1):
    print(f"{' '*(n-i)}{i*'*'}")
```

```python
n=int(input("Enter the size of triangle"))

for i in range(1,n+1):
    print(f"{' '*(n-i)+((2*i-1)*'*')}")
```

**7. Create a calculator app using if-else.**

```python
num1=int(input("Enter the first number :"))
num2=int(input("Enter the second number :"))
op=input("Select the operator to perform task :\n1.'+'\n2.'-'\n3.'*'\n4.'/'\n5.'%'")
res=0

if op=='1' or op=='+':
    res=num1+num2
elif op=='2' or op=='-':
    res=num1-num2
elif op=='3' or op=='*':
    res=num1*num2
elif op=='4' or op=='/':
    if num2!=0:
        res=num1/num2
    else:
        print(f"Denominator can't be zero for performing this operation")
elif op=='5' or op=='%':
    if num2!=0:
        res=num1%num2
    else:
        print(f"Denominator can't be zero for performing this operation")
print(f"Result = {res}")
```

```python
decimal=int(input("Enter a decimal number :"))
num=decimal
l=[]
binary=[]
while num>0:
    l.append(num%2)
    num//=2
n=len(l)    
# for i in range(len(l)):
binary.extend(l[::-1])
print(f"The Binary value of {decimal} is : {binary}")
```

**12. Write a program to find the sum of all even numbers in a list.**

```python
num=int(input("Enter The numbers :"))
l=[]    
res=0
while num>0:
    l.append(num%10)
    num=num//10
for i in l:
    if i%2==0:
        res+=i
print(f"Sum of all Even numbers in the list {l} is : {res}")
```

**13. Create a program to check whether a number is prime or not.**

```python
num=int(input("Enter The numbers :"))
   
if num==1:
    print(f"{num} is a prime number")
else:
    for i in range(2,num):
        if num%i==0:
            print(f"{num} is not a Prime Number")
            break
    else:
        print(f"{num} is a Prime number")
```

**14. Write a program to display the cube of the number up to an integer.**

```python
num=(input("Enter the end number :"))
num=num[::-1]
num=int(num)
l=[]

while num>0:
    l.append(num%10)
    num=num//10

for i in range(len(l)):
    print(f"Cube of {l[i]} = {l[i]**3}")
```
