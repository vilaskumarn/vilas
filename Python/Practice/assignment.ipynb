```python
word="hello world"
l=list(word)
l.remove(" ")
disc={}
t=tuple(l)
t
```

```python
for i in range(len(t)):
    count=t[0:i+1].count(t[i])
    value=count
    disc[t[i]]=value
```

```python
disc
```

```python
for i in range(len(t)):
    count=t.count(t[i])
    # value=count
    disc[t[i]]=count
```

Sort by salary, then experience

Filter high salary but low experience

```python
employees = [
    ("John", 50000, 5),   # (name, salary, experience)
    ("Alice", 70000, 3),
    ("Bob", 50000, 7)
]
```

```python
salary=sorted(employees,key= lambda x: x[1])
salary
```

```python
experience=sorted(employees,key=lambda x:x[2])
experience
```

using **filter()** for the previously sorted variable **experience** gives the list of people with high salary and low experience

```python
high_salary_low_exp=list(filter(lambda x: x[2],experience))
```

```python
high_salary_low_exp
```
