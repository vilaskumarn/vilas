### Logical Operators

Logical operators are used to combine multiple conditions..
They operate on Boolean values and return True or False

**and** : Returns True if both conditions are True

**or** : Returns True if At AT LEAST ONE. condition is True

**not** : Reverse the Boolean value..



```python
age = 25
has_id = True
```

```python
# and

print(age >= 18 and has_id)
```

```python
age = 10
has_id = True
print(age >= 18 and has_id)
```

```python
print(age >= 18 and has_id and age < 10)
```

```python
# or
age = 25
has_id = True
print(age < 18 or has_id)
```

```python
has_id
```

```python
## not
print(not has_id)
```

## Identity Operators..

Identity operators check whether two variables refer to the same object in memory..


This is different from == which checks value equality..


**is** : True if both variables point to the same object

**is not** : True if variables point to different objects..

```python
x = [1,2,3]     # [list]
y = x
z = [1,2,3]
```

```python
x
```

```python
y
```

```python
z
```

```python
x = [1,2,3]     # [list]
y = x
z = [1,2,3]

print(x is y)
print(x is z)
print(x == z)
print(x is not z)
print(x is not y)
```

### Membership Operators..

Membership operators test whether a value is a memeber of sequence..

**in** : Returns True if value if found in the sequence

**not in** :  Returns True if value is NOT found..

```python
fruits = ["apple", "banana", "cherry"]
```

```python
print("apple" in fruits)
print("Apple" in fruits)
print("mango" in fruits)
print('mango' not in fruits)
```

```python
numbers = [1,2,3,4]
print(1 in numbers)
print(1 not in numbers)
```
