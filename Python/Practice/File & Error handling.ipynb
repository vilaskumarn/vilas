**1. Basic file handling**

In Python, the main function is **open()** which takes a filename and a mode like "r" or "w".

**Common modes:**

"r": read (file must exist)

"w": write (creates file, overwrites if it exists)

"a": append (adds at end, creates if not exists)

"r+": read and write

**Very typical patterns:**

*Read all text:*

f = open("data.txt", "r")

content = f.read()

f.close()

*Write text:*

f = open("data.txt", "w")

f.write("Hello\n")

f.close()

**Example A: Write then read**

```python
# create or overwrite a file

f=open("Data.txt","w")      # 'w' = write mode
f.write("First Line\n")
f.write("Second Line")
f.close()
```

```python
# now read it back

f=open("Data.txt","r")
content=f.read()
print(content)
f.close()
```

```python
f=open("Data.txt",'w')
f.write("This is the new text which is being overwritten\n")
f.close()
```

```python
f=open("Data.txt",'a')
f.write("This is the second new text which is being appended\n")
f.close()
```

**2) Using with for automatic closing**

```python
with open("data.txt", "r") as f:
    content = f.read()
    print("Inside with block:")
    print(content)

print("Outside with block: file is already closed here.")
```

**3. File modes in detail**

***When you do open("file.txt", "mode"), the mode controls what you can do.***

Common text modes:

- "r": read only, error if file doesn’t exist

- "w": write (truncate/overwrite, create if not exists)

- "a": append (add at end, create if not exists)

- "r+": read and write, file must exist

- "w+": read and write, truncates file or creates new

- "a+": read and append, file pointer starts at end

- Add "b" for binary (e.g. "rb", "wb") for images, PDFs, etc.

```python
f = open("modes_demo.txt", "w")
print("Writable?", f.writable())
print("Readable?", f.readable())
f.close()
```

```python
f = open("modes_demo.txt", "w+")
print("Writable?", f.writable())
print("Readable?", f.readable())
f.close()
```

```python
f = open("modes_demo.txt", "r")
print("Writable?", f.writable())
print("Readable?", f.readable())
f.close()
```

```python
f = open("modes_demo.txt", "r+")
print("Writable?", f.writable())
print("Readable?", f.readable())
f.close()
```

```python
f = open("modes_demo.txt", "a")
print("Writable?", f.writable())
print("Readable?", f.readable())
f.close()
```

```python
f = open("modes_demo.txt", "a+")
print("Writable?", f.writable())
print("Readable?", f.readable())
f.close()
```

**4.Reading files in different ways**

```python
#read() – whole file or part of it

with open("data.txt", "r") as f:
    whole = f.read()          # read entire file
    print("Whole file:")
    print(whole)
```

```python
#You can pass a size limit:

with open("data.txt", "r") as f:
    first_10 = f.read(10)     # first 10 characters
    print("First 10 chars:", first_10)
```

```python
# readline() – one line at a time

with open("data.txt", "r") as f:
    line1 = f.readline()
    line2 = f.readline()
    # line3=f.readline()
    print("Line 1:", line1)
    print("Line 2:", line2)
    # print("Line 3:", line3)
```

```python
# readlines() and looping

with open("data.txt", "r") as f:
    lines = f.readlines()   # list of lines
    for line in lines:
        print("Line:", line.strip())
```

```python
# readlines() and looping

with open("sample.txt", "r") as f:
    lines = f.readlines()   # list of lines
    for line in lines:
        print("Line:", line.strip())
```

```python
# Even better (no big list in memory):

with open("sample.txt", "r") as f:
    for line in f:          # file object is iterable
        print("Line:", line.strip())
```

**5.Writing and appending**

```python
# write() – one string

with open("notes.txt", "w") as f:
    f.write("Line 1\n")
    f.write("Line 2\n")
```

```python
# writelines() – many strings at once
# You must include \n yourself.

lines = ["First\n", "Second\n", "Third\n"]

with open("notes2.txt", "w") as f:
    f.writelines(lines)
```

```python
with open("notes2.txt",'r') as f:
    content=f.read()
    print(content)
```

```python
lines = ["Fourth\n", "Fifth\n", "Sixth\n"]

with open("notes2.txt", "a") as f:
    f.writelines(lines)
```

```python
with open("notes2.txt",'r') as f:
    content=f.read()
    print(content)
```

**6.Moving inside a file: seek and tell**

Files have a “cursor” (current position). You can move it around.

- tell() → current position (number of bytes/characters from start)

- seek(offset, whence) → move cursor

- whence = 0 from start (default)

- whence = 1 from current position

- whence = 2 from end

```python
# Example: partial read and seek

with open("cursor_demo.txt", "w") as f:
    f.write("ABCDEFGHIJ")   # 10 characters

with open("cursor_demo.txt", "r") as f:
    print("Start position:", f.tell())   # usually 0
    first_four = f.read(4)
    print("After reading 4:", f.tell())
    print("First four:", first_four)

    f.seek(0)               # go back to start
    print("After seek(0):", f.tell())
    all_data = f.read()
    print("All data:", all_data)

    # f.seek(0)

    print(f.tell())
    # f.seek(-3,2)
    # f.seek(-3, 2)           # 3 chars before end
    # last_three = f.read()
    # print("Last three:", last_three)
```

**7.Extracting images from a PDF (slightly advanced)**

**Simple extraction pattern:**

```python
import fitz  # PyMuPDF

pdf = fitz.open("file.pdf")

for page_index in range(len(pdf)):
    page = pdf[page_index]
    images = page.get_images(full=True)

    for image_index, img_info in enumerate(images, start=1):
        xref = img_info[0]
        base_image = pdf.extract_image(xref)
        image_bytes = base_image["image"]
        image_ext = base_image["ext"]
        image_name = f"page{page_index+1}_img{image_index}.{image_ext}"

        with open(image_name, "wb") as img_file:
            img_file.write(image_bytes)
        print("Saved:", image_name)
```
