# 1. Multiplication Table
for i in range(1, 6):
    for j in range(1, 6):
        print(i * j, end=" ")
    print()

# 2. Star Pattern (Rectangle)
for i in range(5):
    for j in range(5):
        print("*", end=" ")
    print()

# 3. Right Triangle Pattern
for i in range(5):
    for j in range(i + 1):
        print("*", end=" ")
    print()

# 4. List of Pairs
nums = [1, 2, 3]
for i in nums:
    for j in nums:
        print(i, j)

# 5. Matrix Addition
A = [[1,2],[3,4]]
B = [[5,6],[7,8]]
C = [[0,0],[0,0]]

for i in range(2):
    for j in range(2):
        C[i][j] = A[i][j] + B[i][j]

print(C)

# 6. Finding Pairs with Sum
arr = [1, 2, 3, 4]
target = 5

for i in range(len(arr)):
    for j in range(i+1, len(arr)):
        if arr[i] + arr[j] == target:
            print(arr[i], arr[j])

# 7. Compare Two Lists
a = [1, 2, 3]
b = [3, 2, 1]

for i in a:
    for j in b:
        if i == j:
            print("Match:", i)

# 8. 2D Grid Coordinates
for x in range(3):
    for y in range(3):
        print(f"({x},{y})")

# 9. Frequency Count (Basic)
arr = [1, 2, 2, 3, 3, 3]

for i in arr:
    count = 0
    for j in arr:
        if i == j:
            count += 1
    print(i, "->", count)

# 10. Nested Loop with Conditions
for i in range(1, 4):
    for j in range(1, 4):
        if i != j:
            print(i, j)
