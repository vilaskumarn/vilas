# Maths operations on set..



## UNION -- combine two Sets (all unique items)

**mehotds .union()**\
return all unique items from both sets

```python
A={2,3,5,8}
B={1,4,6,7}

result=A.union(B)

print(f"Result of Set A Union Set B : {result}")
```

**Union with multiple sets**

```python
# We can combine multiple cells using UNION operator '|'

C={8,9,10,11}
result = A | B | C

print(f"Result of Set A Union B Union C : {result}")
```

## Intersection - Common item only
**& and methos .intersection()**\
Return only items that exists in both the sets

```python
A={2,3,5,7,8}
B={2,4,5,7,9}

result=A.intersection(B)

print(f"Result of Set A Intersection Set B : {result}")
```

**Intersection of multiple sets**

```python
# We can intersect multiple cells using UNION operator '&'
A={2,3,5,7,8}
B={2,4,5,7,8}
C={8,9,5,7}
result=A&B&C
print(f"Result of Set A & B & C : {result}")
```

## Difference - Items in first but not in second

**Return items in hte first set that are not in 2nd set**\
.difference(), and  " - "

```python
A={1,2,3,5}
B={4,5,7,9}

result=A.difference(B)

print(f"Result of Set A Difference Set B : {result}")
```

```python
A={1,2,3,5}
B={4,5,7,9}

result=A-B

print(f"Result of Set A Difference Set B : {result}")
```

## Symmetric Difference - Item in Either set but not both
**symbol ^ and .symmetic_difference()**\
Return items that are in either setm, but not in both

```python
A={1,2,3,5}
B={4,5,7,9}

result=A^B

print(f"Result of Set A Symmetric Difference Set B : {result}")
```

```python
A={1,2,3,5}
B={4,5,7,9}

result=A.symmetric_difference(B)

print(f"Result of Set A Symmetric Difference Set B : {result}")
```
